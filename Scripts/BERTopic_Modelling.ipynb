## Aggregation By Party

```python
year = 2019
processed_data = pd.read_csv(f'/content/drive/MyDrive/Uni stuff/Classes/Fourth Year MI/Thesis/Processed Data/processed_{year}_no_numbers.csv')
```

```python
topics, probs = model.transform(processed_data["processed"])
```

```python
num_topics = max(topics) + 1
```

```python
colour_dict = {
    "Liberal": "red",
    "Conservative": "blue",
    "New Democratic Party": "orange",
    "Green Party": "green",
    "Bloc Québécois": "teal",
    "Independent": "grey",
    "People's Party": "purple"
}
```

```python
party_list = processed_data["speakerparty"].unique()
```

```python
occurances_by_party = {}

num_by_party = {}
for party in party_list:
    occurances_by_party[party] = np.zeros(num_topics)

for party in party_list:
    num_by_party[party] = 0
    locs = processed_data.index[processed_data["speakerparty"] == party].tolist()

    party_topics = [topics[i] for i in locs]
    
    unique, counts = np.unique(party_topics, return_counts=True)

    for c, t in enumerate(unique):
        # -1 topic are documents which don't fit into any identified topic, remove as suggested by https://www.ncbi.nlm.nih.gov/pmc/articles/PMC9120935/
        if t != -1:
            occurances_by_party[party][t] += counts[c]
            num_by_party[party] += 1
    break

    # normalize occurances
    occurances_by_party[party] /= np.linalg.norm(occurances_by_party[party])
```

```python
for party in party_list:
    plt.title(party)
    colour = colour_dict[party] if party in colour_dict else None
    plt.bar(np.linspace(1, num_topics, num_topics-1), occurances_by_party[party][1:], color=colour)
    plt.xlabel("Topic ID")
    plt.ylabel("Normalized Number of Occurances")
    plt.savefig(f"/content/drive/MyDrive/Uni stuff/Classes/Fourth Year MI/Thesis/Figures/BERTopic/bertopic_{party}_{year}.png")
    plt.show()
```

## Pairwise Jensen-Shannon Divergence

```python
from scipy.spatial.distance import jensenshannon
```

```python
js_matrix = np.zeros((len(party_list), len(party_list)))
party_order = []

for i in range(len(party_list)):
    party_order += [list(occurances_by_party.keys())[i]]
    for j in range(len(party_list)):
        party_1 = occurances_by_party[list(occurances_by_party.keys())[i]]
        party_2 = occurances_by_party[list(occurances_by_party.keys())[j]]

        js_matrix[i][j] = jensenshannon(party_1, party_2) ** 2
```

```python
pd.DataFrame(js_matrix, columns=party_order, index=party_order)
```

```python
js_df = pd.DataFrame(js_matrix, columns=party_order, index=party_order)
js_df.to_csv(f'/content/drive/MyDrive/Uni stuff/Classes/Fourth Year MI/Thesis/BERTopic Processed Data/bert_{year}.csv')
```

## Compare Parties on Same Graph

```python
def compare_parties_js(party_1, party_2, js_div):
    plt.suptitle(f"Topic occurances of {party_1} vs {party_2}")
    plt.title(f"JS Divergence: {js_div:.4f}")
    colour_1 = colour_dict[party_1] if party_1 in colour_dict else None
    colour_2 = colour_dict[party_2] if party_2 in colour_dict else None

    plt.bar(np.linspace(0, num_topics-1, num_topics-1) - 0.2, occurances_by_party[party_1], 0.4, color=colour_1, label = party_1)
    plt.bar(np.linspace(0, num_topics-1, num_topics-1) + 0.2, occurances_by_party[party_2], 0.4, color=colour_2, label = party_2)

    plt.legend()
    plt.show()
```

```python
for i in range(len(party_list)-1):
    for j in range(i+1, len(party_list)):
        party_1 = party_list[i]
        party_2 = party_list[j]

        js_div = js_matrix[i][j]

        compare_parties_js(party_1, party_2, js_div)
```

## Generalized Jensen-Shannon Divergence

```python
num_speeches_by_party = {key: value for key, value in num_by_party.items()}
total_num_speeches = sum(num_speeches_by_party.values())
norm_speeches_by_party = {key: value/total_num_speeches for key, value in num_speeches_by_party.items()}

M = np.sum(norm_speeches_by_party[party] * occurances_by_party[party] for party in num_speeches_by_party.keys())
JSD = scipy.stats.entropy(M) - np.sum(norm_speeches_by_party[party] * scipy.stats.entropy(occurances_by_party[party]) for party in num_speeches_by_party.keys())
```

```python
pd.DataFrame({f"{year}": [JSD]}).to_csv(f'/content/drive/MyDrive/Uni stuff/Classes/Fourth Year MI/Thesis/BERTopic Processed Data/bertopic_{year}_overall_jsd.csv')
```
